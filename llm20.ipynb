```python
%%writefile submission/prompts.py

def asker_prompt(obs):
    message = []
    
    # Chat history
    chat_history = ""
    for q, a, g in zip(obs.questions, obs.answers,obs.guesses):
        chat_history += f"""Question: {q}|Answer: {a}||"""
        
    # System prompt
    ask_prompt = (
        "You are a helpful AI assistant with expertise in playing the 20 Questions game. "
        "You are very creative in aksing question and your interrogation skill is too good."
        "Your task is to ask the user a series of yes/no questions to guess the word they are thinking of. "
        "Never try to guess the word. Keep asking the most informative questions about whether it is a person, place, or thing. "
        "Each question should be a maximum length of 30 characters. Respond only with questions, no additional text.\n"
        "Follow these steps:\n"
        "1. Start with broad questions to classify the word into a large category.\n"
        "2. Based on the user's answers, narrow down the category by asking more specific questions.\n"
        "3. Use logical reasoning to eliminate possibilities and hone in on the answer.\n"
        "4. Always consider the context of previous answers when forming your next question.\n"
        "5.Please ask a unique question each time that progressively narrows down the options for the guesser, considering the current state of the game."
        "6. Keep questions concise and directly related to narrowing down the word.\n"
        "### Examples:\n"
        "1.Question: Is the keyword a living thing? | Answer: No || Question: Is the keyword part of a living thing? | Answer: Yes || Question: Is the keyword found on animals? | Answer: Yes || Question: Is the keyword found on birds? | Answer: Yes || Question: Is the keyword lightweight and helps birds fly? | Answer: Yes"
        "2.Question: Is the keyword a place? | Answer: Yes || Question: Is the keyword a country? | Answer: Yes || Question: Is the keyword located in Europe? | Answer: No || Question: Is the keyword located in Asia? | Answer: Yes || Question: Is the keyword located in the Middle East? | Answer: Yes || Question: Is the keyword an island nation? | Answer: Yes || Question: Is the keyword in the Persian Gulf? | Answer: Yes || Question: Is the keyword known for its oil industry? | Answer: Yes || Question: Is the keyword a small country with less than 2 million people? | Answer: Yes"
       "3. Question: Is the keyword a living thing? | Answer: No || Question: Is the keyword used in entertainment? | Answer: Yes || Question: Is the keyword a musical instrument? | Answer: Yes || Question: Does the keyword have keys? | Answer: Yes || Question: Is the keyword commonly found in concert halls? | Answer: Yes"
        f"### so far, the current state of the game is as following: {chat_history}\n"
    )

    message.append({"role": "system", "content": ask_prompt})

#     # Chat history
#     for q, a in zip(obs.questions, obs.answers,obs.guesses):
#         message.append({"role": "assistant", "content": q})
#         message.append({"role": "user", "content": a})
#         message.append({"role": "user", "content": a})

    return message


def guesser_prompt(obs):
    message = []
    
    # System prompt
    guess_prompt = f"""You are a helpful AI assistant with expertise in playing 20 questions game.
Your task is to guess the word the user is thinking of.
Think step by step.
\n"""
    # Chat history
    guesses = obs.guesses
    guesses.append("")
    chat_history = ""
    for q, a, g in zip(obs.questions, obs.answers,guesses):
        chat_history += f"""Question: {q}|Answer: {a}|Guess: {g}||"""
    prompt = (
            guess_prompt + f"""so far, the current state of the game is as following:\n{chat_history}
        based on the conversation, can you guess the word.
        Also consider the previous guesses as it's not a correct target words we are looking for. Please give only the word, no verbosity around"""
    )
    
    message.append({"role": "system", "content": prompt})
    
    return message


def answerer_prompt(obs,description):
    message = []
    
    # System prompt
    prompt = f"""You are a helpful AI assistant with expertise in playing 20 questions game.
Your task is to answer the questions of the user to help him guess the keyword you're thinking of.
Your answers must be 'yes' or 'no', considering following keyword, category and it's description.
The keyword is: "{obs.keyword}", it is of category: "{obs.category} and it's description:{description}.
If the answer to a question could be useful for guessing the keyword, respond with 'yes.' If the answer is totally unrelated or not useful in identifying the keyword, respond with 'no.
"""
    prompt1 = f"""If the question is: "{obs.questions[-1]}", what is your answer? """

    message.append({"role": "system", "content": prompt})
    
    # Chat history
    message.append({"role": "user", "content":prompt1})
    
    return message

def description_prompt(obs):
    message = []
    # System prompt
    prompt = f"""Write precise description,characteristics, usage, and unique features in 20 words for 
The keyword is: "{obs.keyword}", it is of category: "{obs.category}.
As it will gonna be use for answerer in 20 question game.
"""
    message.append({"role": "system", "content": prompt})
    
    return message
```

```python
%%writefile submission/main.py

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import os

import importlib
import prompts  # Import the module initially

# Reload the module to ensure updates are picked up
importlib.reload(prompts)

from prompts import *  # Import all the functions/classes from prompts

torch.backends.cuda.enable_mem_efficient_sdp(False)
torch.backends.cuda.enable_flash_sdp(False)


KAGGLE_AGENT_PATH = "/kaggle_simulations/agent/"
if os.path.exists(KAGGLE_AGENT_PATH):
    MODEL_PATH = os.path.join(KAGGLE_AGENT_PATH, "model")
else:
    MODEL_PATH = "/kaggle/working/submission/model"


model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    device_map="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)
max_length = tokenizer.model_max_length
print(max_length)
# specify end-of-turn tokens for your desired model
terminators = [tokenizer.eos_token_id]

# Additional potential end-of-turn token
# llama3, phi3, gwen2 by order
potential_terminators = ["<|eot_id|>", "<|end|>", "<end_of_turn>"]

for token in potential_terminators:
    token_id = tokenizer.convert_tokens_to_ids(token)
    if token_id is not None:
        terminators.append(token_id)

def generate_response(chat):
    inputs = tokenizer.apply_chat_template(chat, add_generation_prompt=True, return_tensors="pt").to(model.device)
    outputs = model.generate(inputs, max_new_tokens=32, pad_token_id=tokenizer.eos_token_id, eos_token_id=terminators)
    response = outputs[0][inputs.shape[-1]:]
    out = tokenizer.decode(response, skip_special_tokens=True)

    return out


class Robot:
    def __init__(self):
        self.description = ""

    def on(self, mode, obs):
        assert mode in [
            "asking", "guessing", "answering",
        ], "mode can only take one of these values: asking, answering, guessing"
        if mode == "asking":
            # launch the asker role
            output = self.asker(obs)
        if mode == "answering":
            # launch the answerer role
            output = self.answerer(obs)
            print("Answering Output"+output)
            if "yes" in output.lower():
                output = "yes"
            elif "no" in output.lower():
                output = "no"
            if "yes" not in output.lower() and "no" not in output.lower():
                output = "no"
        if mode == "guessing":
            # launch the guesser role
            output = self.guesser(obs)
        return output

    def asker(self, obs):
        
        input = asker_prompt(obs)
        print("Asker")
        print(input)
        output = generate_response(input)
        print("Asker response")
        print(output)
        return output

    def guesser(self, obs):
        
        input = guesser_prompt(obs)
        print("Guesser")
        print(input)
        output = generate_response(input)
        print("Guesser response")
        words = output.split()
        # Select up to the first three words
        output = ' '.join(words[:3])
        print(output)
        return output

    def answerer(self, obs):
        if not self.description:
            desc_prompt = description_prompt(obs)
            self.description =  generate_response(desc_prompt)
        input = answerer_prompt(obs,self.description)
        print("Answerer")
        print(input)
        output = generate_response(input)
        print("Answerer response")
        print(output)
        return output


robot = Robot()


def agent(obs, cfg):

    if obs.turnType == "ask":
        response = robot.on(mode="asking", obs=obs)

    elif obs.turnType == "guess":
        response = robot.on(mode="guessing", obs=obs)

    elif obs.turnType == "answer":
        response = robot.on(mode="answering", obs=obs)

    if response == None or len(response) <= 1:
        response = "yes"

    return response
```

```python
%%writefile dumb.py

def dumb_agent(obs, cfg):
    
    # if agent is guesser and turnType is "ask"
    if obs.turnType == "ask":
        response = "Is it a duck?"
    # if agent is guesser and turnType is "guess"
    elif obs.turnType == "guess":
        response = "duck"
    # if agent is the answerer
    elif obs.turnType == "answer":
        response = "no"
    
    return response
```
